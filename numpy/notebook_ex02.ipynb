# Exercise 2: Zeros

The goal of this exercise is to learn to create a NumPy array with 0s.

1. Create a NumPy array of dimension **300** with zeros without filling it manually
2. Reshape it to **(3,100)**

```python
import numpy as np

# Create a NumPy array of dimension 300 with zeros
zeros_array = np.zeros(300)
zeros_array
```

```python
# Reshape it to (3, 100)
reshaped_array = zeros_array.reshape(3, 100)
reshaped_array
```
